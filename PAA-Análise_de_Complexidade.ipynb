# Análise de Complexidade 


Abordagem usada para descrever o desempenho e a eficiência de algoritmos em termos de tempo de execução ou uso de recursos (como memória) à medida que o tamanho da entrada aumenta. A notação "Big O" é uma das várias notações usadas na análise de algoritmos e pode ser considerada como um aspecto específico da análise de complexidade.

## Notações de Complexidade
Além do "Big O", existem outras notações como "Big Omega" (Ω) e "Big Theta" (Θ), que fornecem limites inferiores e limites exatos, respectivamente, para a complexidade de um algoritmo.

## Ordens de Complexidade
Além do "O(n)" (linear), existem outras ordens de complexidade comuns, como "O(1)" (constante), "O(log n)" (logarítmica), "O(n log n)" (linearítmica) e "O(n^2)" (quadrática), entre outras.

## Análise de Algoritmos
A análise de complexidade é um tópico importante na análise de algoritmos, que se concentra em avaliar o desempenho dos algoritmos em relação ao tamanho da entrada. Isso inclui a escolha de algoritmos eficientes para resolver problemas específicos.

# Notação "Big O"

A **Notação "Big O"** se refere a uma função matemática aplicada na ciência da computação para analisar a complexidade algorítmica. Ela nos informa sobre o tempo de execução necessário para um determinado algoritmo, mas não nos dirá o quão rápido o algoritmo é em tempo de execução. Em vez disso, ela nos informa como nosso algoritmo se comporta sob *diferentes tamanhos de entrada* (**dinamicamente**).

Se um algoritmo realiza uma operação em cada item de uma matriz de tamanho n, esse algoritmo tem um custo computacional de **O(n)** e tem um *custo computacional de O(1) para cada item*. É importante observar, no entanto, que a notação **"Big O" é definida de acordo com o limite superior do algoritmo**, o que significa que ela é classificada no **pior cenário**.

Existem diferentes classes de Notação "Big O", e essas classes nos ajudam a classificar a **eficiência do algoritmo**.

Abaixo estão cinco das classes de complexidade temporal mais comuns:

### O(1) - Custo Computacional Constante
**O custo de complexidade O(1)** pode ser classificado como um algoritmo que leva a mesma quantidade de tempo para ser concluído, independentemente do tamanho da entrada. Isso significa que o tempo que leva para o algoritmo ser executado é constante, o que indica que **não é afetado pelo tamanho da entrada**.

### O(log n) - Custo Computacional de Complexidade Logarítmica
A complexidade de custo computacional **O(log n)** é exemplificada quando o tempo necessário para a conclusão é proporcional ao logaritmo do tamanho da entrada. **Conforme o tamanho da entrada aumenta, a quantidade de tempo necessária para concluir o algoritmo aumenta, mas a uma taxa mais lenta**.

### O(n) - Custo Computacional de Complexidade Linear
Um algoritmo possui complexidade de **custo computacional O(n)** quando o tempo necessário para a conclusão é diretamente proporcional ao tamanho da entrada. Isso significa que o tempo necessário para concluir o algoritmo **aumenta linearmente à medida que o tamanho da entrada cresce**.

### O(n log n) - Custo Computacional de Complexidade Linearítmica
Um algoritmo possui complexidade de **custo computacional O(n log n)** quando o **tempo necessário para a conclusão é proporcional a n multiplicado pelo logaritmo de n**. Isso é comumente observado em *algoritmos de divisão e conquista*.

### O(n^2) - Custo Computacional de Complexidade Quadrática
Um algoritmo possui complexidade de **custo computacional O(n^2)** quando o tempo necessário para a conclusão é proporcional ao quadrado do tamanho da entrada. **À medida que o tamanho da entrada aumenta, o tempo necessário para concluir o algoritmo aumenta exponencialmente**.

Existem muitas outras complexidades de custo computacional "Big-O", no entanto, esses exemplos são suficientes para demonstrar o que queremos dizer com "Big-O", sua implementação e as diferenças entre diferentes complexidades de custo computacional.

```python
from math import log       # import done only to use logarithm
import numpy as np         # import done to use of calculation, arrays
import matplotlib.pyplot as plt   # import done only to use graphics

# Constant Function
# Created array of 100 position whith values 1 until 10.

n = np.linspace(0.01, 10, 100)     # random value #Linear Function

big_o_const = np.ones(n.shape)  #Constant Function

big_o_log = np.log(n)           #Logarithmic Function

big_o_log_linear = n*big_o_log  #Logarithmic Linear Function

big_o_quadratic = n**2          #Quadratic Function

big_o_cubic = n**3              #Cubic Function

big_o_exponential = 2**n        #Exponential Function

labels = [ 'Constante: O(1)', 'Logarítmica: O(log n)', 'Linear: O(n)',
          'Log Linear: O(n log n)','Quadrática: O(n^2)','Cúbica: Quadrática: O(n^3)','Exponencial: O(2^n)']
big_o = [big_o_const, big_o_log, n, big_o_log_linear, big_o_quadratic, big_o_cubic, big_o_exponential]

# Plot das curvas Big O
plt.figure(figsize=(10, 8))
plt.ylim(0, 100)
plt.xlim(0, 10)
for i in range(len(big_o)):
    plt.plot(n, big_o[i], label=labels[i])
# Criação da primeira caixa de legenda com as labels das curvas Big O
plt.legend()

## CRIAR AS ZONAS DE COMPLEXIDADE E AS LEGENDAS DE CORES
# Preenchimento do fundo com cores e áreas específicas
colors = ['green', 'lightgreen', 'yellow', 'orange', 'red']
# Criação da segunda caixa de legenda com símbolos de cores
legend_labels = ['Excelente', 'Bom', 'Ok', 'Ruim', 'Péssimo']
legend_handles = [plt.Line2D([0], [0], marker='s', color='w', markerfacecolor=color, markersize=10) for color in colors]
# Adicione a segunda legenda acima do gráfico
plt.gca().add_artist(plt.legend(legend_handles, legend_labels, title='Custo de Complexidade Big-O', loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=5))

plt.fill_between(n, 0, 100, color='red', alpha=0.2, interpolate=True)
plt.fill_between(n, 5*n,  color='orange', alpha=0.6, interpolate=True)
plt.fill_between(n, 2*n,  color='yellow', alpha=0.7, interpolate=True)
plt.fill_between(n, 3/4*n,  color='lightgreen', alpha=0.3, interpolate=True)
plt.fill_between(n, n/4,  color='green', alpha=0.33, interpolate=True)
## FIM DAS ZONAS DE COMPLEXIDADE E AS LEGENDAS DE CORES

# Adicionar texto às linhas
plt.text(3.9, 98, 'O(n^3)', fontsize=10, color='blue', ha='left', va='center')
plt.text(5.9, 98, 'O(2^n)', fontsize=10, color='blue', ha='left', va='center')
plt.text(9.2, 98, 'O(n^2)', fontsize=10, color='blue', ha='left', va='center')
plt.text(8.8, 23, 'O(n log(n))', fontsize=10, color='blue', ha='left', va='center')
plt.text(9.6, 12, 'O(n)', fontsize=10, color='blue', ha='left', va='center')
plt.text(9.2, 4, 'O(log(n))', fontsize=10, color='blue', ha='left', va='center')

# Adicionar uma grade
plt.grid(True, linestyle='dotted')
plt.legend(title='Complexidade', loc='upper left')
plt.ylabel('Operações')
plt.xlabel('n')
plt.show()
```
